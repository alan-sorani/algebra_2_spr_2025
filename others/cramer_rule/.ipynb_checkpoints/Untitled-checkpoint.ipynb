```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
scale_first = True
v_1 = np.array([20, 0])
v_2 = np.array([40, 60])
origin = np.array([0,0])
x = np.array([2.2, 1.5])
b = x[0] * v_1 + x[1] * v_2
scale = x[0] if scale_first else x[1]

v_1_scaled = scale * v_1 if scale_first else v_1
v_2_scaled = v_2 if scale_first else scale * v_2

parallelogram_small = np.array([
    origin,
    v_1,
    v_1 + v_2,
    v_2,
    origin
])

parallelogram_scaled = np.array([
    origin,
    v_1_scaled,
    v_1_scaled + v_2_scaled,
    v_2_scaled,
    origin
])

parallelogram_cramer = np.array([
    origin,
    b if scale_first else v_1,
    b + v_2 if scale_first else v_1 + b,
    v_2 if scale_first else b,
    origin
])

parallelogram_big = np.array([
    origin,
    x[0] * v_1,
    x[0] * v_1 + x[1] * v_2,
    x[1] * v_2,
    origin    
])
```

```python
ax = plt.gca()
ax.set_xlim(-50, 200)
ax.set_ylim(-50, 300)

# parallelogram with the scaled vector replaced by b
ax.plot(parallelogram_cramer[:, 0], parallelogram_cramer[:, 1], color='blue')
ax.fill(parallelogram_cramer[:, 0], parallelogram_cramer[:, 1], color='blue', alpha=0.2)

# scaled parallelogram
ax.plot(parallelogram_scaled[:, 0], parallelogram_scaled[:, 1], color='green')
ax.fill(parallelogram_scaled[:, 0], parallelogram_scaled[:, 1], color='green', alpha=0.2)

# small parallelogram
ax.plot(parallelogram_small[:, 0], parallelogram_small[:, 1], color='red')
ax.fill(parallelogram_small[:, 0], parallelogram_small[:, 1], color='red', alpha=0.2)

# parallelogram with both axes scaled
ax.plot(parallelogram_big[:, 0], parallelogram_big[:, 1], linestyle="dotted", color='black')


# vector names
ax.text(b[0], b[1], 'b', color='blue', fontsize=12, weight='bold', va='top')

'''
ax.text(v_1[0], v_1[1], 'v_1', color='red', fontsize=12, weight='bold', va='top')
ax.text(v_2[0], v_2[1], 'v_2', color='red', fontsize=12, weight='bold', va='bottom')

if(scale_first):
    ax.text(v_1_scaled[0], v_1_scaled[1], f'{x[0]}*v_1', color='green', fontsize=12, weight='bold', va='top')
else:
    ax.text(v_2_scaled[0], v_2_scaled[1], f'{x[1]}*v_2', color='green', fontsize=12, weight='bold', va='bottom')
'''

# vector arrows
plt.arrow(0,0,v_1[0],v_1[1])
    
plt.show()
```
